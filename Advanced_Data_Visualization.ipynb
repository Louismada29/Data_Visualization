```python
# path
path_1 = '/content/gdrive/MyDrive/Dataset & Modelling/order_details.csv'
path_2 = '/content/gdrive/MyDrive/Dataset & Modelling/order_lists.csv'

# load dataset
details = pd.read_csv(path_1)
lists = pd.read_csv(path_2)
```

```python
details.head()
```

```python
lists.head()
```

```python
# join table
dataset = pd.merge(details, lists, how='right', on='order_id')
dataset.head()
```

```python
# image
img = plt.imread('https://www.google.co.id/images/branding/googlelogo/2x/googlelogo_color_160x56dp.png')
plt.imshow(img)
plt.show()
```

```python
# apakah hubungan biaya dengan profit
sns.regplot(x='profit', y='Cost', data= dataset)
plt.show()
```

```python
# menghitung negara yang melakukan order paling tinggi
sns.countplot(data = dataset, y='country')
```

```python
plt.figure(figsize= (18,12))
plt.subplot(2,2,1)
plt.title('Transactions in countries')
sns.countplot(data = dataset, y='country')

plt.subplot(2,2,2)
plt.title('Cost and Profit Correlation')
sns.regplot(x='profit', y='Cost', data= dataset)

plt.subplot(2,2,3)
plt.title('Quantity and sales correlation')
sns.scatterplot(data = dataset,x='sales', y='quantity', hue='category')
plt.xlabel('Our Sales')
plt.ylabel('Our Quantity')
plt.show()
```

```python
sns.countplot(data =dataset, y='city', hue = 'category', order = dataset.city.value_counts().iloc[:10].index
              , palette = ['#312376','grey','grey'])
plt.show()
```
